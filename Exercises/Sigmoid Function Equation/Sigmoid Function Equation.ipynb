# <center>Sigmoid Function Equation</center>

The Logistic Regression Equation/ Sigmoid Function is given by:

<img src='Logistic Regression Equation.png' width=500>

S<sub>w,b</sub>(x) = $\frac{1}{1\ +\ e^-(wx+b)}$   
where,  
w = m = slope of the curve  
b = c = y-intercept

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
def Linear(x, w, b):
    return w*x + b


def Logistic(x, w, b):
    return 1/(1 + np.exp( -(Linear(x, w, b)) ))
```

```python
Linear(10, 2, 1)
```

```python
Logistic(10, 2, 1)
```

### Test plot

```python
w = 1
b = 1

X = np.linspace(-10, 10, 100)
y = Logistic(X, w, b)

plt.plot(X, y)
plt.title("Sigmoid Curve")
plt.show()
```
